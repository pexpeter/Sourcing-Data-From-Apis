# RANDOM USER API USAGE FOR DATA SOURCE

## Importing Python Packages

```python
import pandas as pd
import requests
```

## Making an API Call

We will be generating user data from [random user api](https://randomuser.me/api) and reading it to a panda dataframe. The data from the api is in json format/ dictionaries, and we will be interested in the key/value pair of `results` .




```python
r=requests.get("https://randomuser.me/api/?results=300")
x=r.json()
x
x["results"]
df=pd.DataFrame(x["results"])
df.head(10)
```

## Data Frame Properties

```python
df.info()
```

The data frame has 300 user information as we requested in our api call.

### OFFICIAL NAMES

```python
df["name"][0]
```

```python
df["title"]= [d['title'] for d in df["name"]]
df["title"]
df["official_name"] = [d['first'] + ' ' + d['last'] for d in df["name"]]
df["official_name"]
```

The names are in form of a dictionary with three subcategories we merge the information into two columns
- ``official_name``
- ``title``


```python
df.head()
```

### Location of the Users

```python
df["location"][0]
```

```python
df['street_no'] = [d["street"]["number"] for d in df["location"]]
df["street_name"] = [d["street"]["name"] for d in df["location"]]
df["city"] = [d["city"] for d in df["location"]]
df["state"] = [d["state"] for d in df["location"]]
df["country"] = [d["country"] for d in df["location"]]
df["postcode"] = [d["postcode"] for d in df["location"]]
df["lattitude"] = [d["coordinates"]["latitude"] for d in df["location"]]
df["longitude"] = [d["coordinates"]["longitude"] for d in df["location"]]
df["timezone"] = [d["timezone"]["offset"] for d in df["location"]]
df["gmtcity"] = [d["timezone"]["description"] for d in df["location"]]
```

The location details are in form of a dictionary with six categories and six  subcategories we merge the information into ten columns:
- ``street number``
- ``street name``
- ``city``
- ``state``
- ``country``
- ``postcode``
- ``lattitude``
- ``longitude``
- ``timezone``
- ``timezone city``
